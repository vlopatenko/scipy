```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

```python
# Будем рассматривать экспоненциальное распределение
expon_rv = sts.expon()
```

```python
# Генерируем выборку объема n = 1000 
N = 1000
expon_sample = expon_rv.rvs(N)
```

```python
# Генерируем сетку
x = np.linspace(expon_rv.ppf(0.01), expon_rv.ppf(0.99), N)

# Рисуем гистограмму выборки и теоретическую плотность поверх нее
fig, ax = plt.subplots(1, 1)

ax.hist(expon_sample, normed=True, bins=18)
ax.plot(x, expon_rv.pdf(x))

ax.set_xlabel('$x$')
ax.set_ylabel('$fraction of samples$')
```

## Генерация 1000 выборок объема различных объемов n

```python
# Заранее определим функцию, которая будет принимать на вход n и возвращать выборочные средние для выборок объема n
def samples_average(n):
    samples = np.array([sts.expon().rvs(n) for _ in range(N)])
    # Считаем средние для каждой выборки
    return np.mean(samples, axis=1)
```

# n = 5

```python
n = 5
# Генерируем 1000 выборок объема 5
samples_avg = samples_average(n)

# Считаем теоретические значения параметров распределения
n_mean = expon_rv.mean()
n_std = expon_rv.std() / np.sqrt(n)

# Генерируем гаусовскую случайную величину
gaussian_rv = sts.norm(n_mean, n_std)
x = np.linspace(gaussian_rv.ppf(0.01), gaussian_rv.ppf(0.99), N)

# Отрисовываем гистограмму и плотность нормального распределения поверх
fig, ax = plt.subplots(1, 1)
ax.hist(samples_avg, normed=True, bins=18)
ax.plot(x, gaussian_rv.pdf(x))
ax.set_xlabel('$x$')
ax.set_ylabel('$fraction of samples$')
```

# n = 10

```python
# Как видно, распределение средних уже при n = 5 напоминает нормальное, оно унимодальное, похожее по форме, однако несколько
# угловатое и скошено. Повторим процедуру для n = 10
n = 10
# Генерируем 1000 выборок объема 10
samples_avg = samples_average(n)

# Считаем теоретические значения параметров распределения
n_mean = expon_rv.mean()
n_std = expon_rv.std() / np.sqrt(n)

# Генерируем гаусовскую случайную величину
gaussian_rv = sts.norm(n_mean, n_std)
x = np.linspace(gaussian_rv.ppf(0.01), gaussian_rv.ppf(0.99), N)

# Отрисовываем гистограмму и плотность нормального распределения поверх
fig, ax = plt.subplots(1, 1)
ax.hist(samples_avg, normed=True, bins=18)
ax.plot(x, gaussian_rv.pdf(x))
ax.set_xlabel('$x$')
ax.set_ylabel('$fraction of samples$')
```
